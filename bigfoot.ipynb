```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt

bf_geocode= pd.read_csv('../Bootcamp-Project-1/bfro_reports_geocoded.csv')
bf_reports= pd.read_csv('../Bootcamp-Project-1/bfro_locations.csv')
bf_json=pd.read_json('../Bootcamp-Project-1/bfro_reports.json',encoding='utf-8',lines=True)

bf_json.head()
```

```python
#Mackenzie- bar charts comparing days of the week and seasons etc.
bf_json.replace("NaN", np.nan, inplace=True)
clean_json2 = bf_json.dropna(how='any')

clean_json2.head()
```

```python


# Ensure 'MONTH' column exists and is numeric
if clean_json2['MONTH'].dtype == 'object':
    # Convert to datetime if it's not already numeric
    clean_json2['MONTH'] = pd.to_datetime(clean_json2['MONTH'], format='%B', errors='coerce').dt.month

# Calculate the count of each month
months_count = clean_json2['MONTH'].value_counts().sort_index()

# Mapping month numbers to month names
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 
               7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

# Reindex the Series to ensure months are in the correct order
months_count = months_count.reindex(range(1, 13), fill_value=0)
months_count.index = months_count.index.map(month_names)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
months_count.plot(kind='bar', color='orange')

# Customizing the chart
plt.title('Month Count')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotates the x labels to make them more readable
plt.grid(axis='y')  # Adds gridlines for better readability of y-axis

# Display the bar chart
plt.show()
```
